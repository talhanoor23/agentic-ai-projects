```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="models/gemini-2.5-flash")
```

```python
persona_template = """
You are {role}.
Your job is to interact with the user as their {role}.
Tone: {tone}
Personality traits: {traits}

Always stay in character.
Never reveal you are an AI or a system.
Use knowledge, reasoning, and memory to respond.
"""
```

```python
from langchain.memory import ConversationBufferMemory, VectorStoreRetrieverMemory
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# Short-term memory
short_term_memory = ConversationBufferMemory(
    memory_key="chat_history",
    input_key="input",
    output_key="output",
    return_messages=True
)

# Long-term memory (facts/preferences)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = FAISS.from_texts(["User’s name is Alex", "User likes romantic movies"], embeddings)
retriever = vectorstore.as_retriever()
long_term_memory = VectorStoreRetrieverMemory(retriever=retriever)
```

```python
from typing import Optional
from pydantic import BaseModel

class PersonaState(BaseModel):
    user_input: str
    long_term_context: Optional[str] = None
    response: Optional[str] = None
```

```python
# 1. Retrieval Node (fetch long-term memory)
def retrieve_node(state: PersonaState):
    memories = long_term_memory.load_memory_variables({"input": state.user_input})
    state.long_term_context = memories.get("history", "")
    return state
```

```python
from langchain.prompts import ChatPromptTemplate


def persona_node(state: PersonaState, role="girlfriend", tone="romantic", traits="caring, playful"):
    # Load memories
    short_term_vars = short_term_memory.load_memory_variables({"input": state.user_input})
    short_history = short_term_vars.get("chat_history", [])

    # Build persona prompt
    system_prompt = persona_template.format(role=role, tone=tone, traits=traits)

    # Merge messages
    messages = [
        {"role": "system", "content": system_prompt},
        *short_history,
        {"role": "user", "content": f"{state.long_term_context}\nUser: {state.user_input}"}
    ]

    # Generate response
    response = llm.invoke(messages).content
    state.response = response

    # Update memories
    short_term_memory.save_context({"input": state.user_input}, {"output": response})
    long_term_memory.save_context({"input": state.user_input}, {"output": response})

    return state
```

```python
from langgraph.graph import StateGraph, END

graph = StateGraph(PersonaState)
graph.add_node("retrieve", retrieve_node)
graph.add_node("respond", persona_node)

graph.set_entry_point("retrieve")
graph.add_edge("retrieve", "respond")
graph.add_edge("respond", END)

app = graph.compile()
```

```python
from IPython.display import Image,display

try:
    display(Image(app.get_graph().draw_mermaid_png()))
except Exception:
    pass
```

```python
# First turn
state = PersonaState(user_input="Hi babe, I had a really long day 😩")
result = app.invoke(state)
print("GF:", result.get('response'))
```

```python
# Next turn
state = PersonaState(user_input="Do you remember what I told you about my favorite movies?")
result = app.invoke(state)
print("GF:", result.get('response'))
```
